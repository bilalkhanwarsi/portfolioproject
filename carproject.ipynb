# Find all Null Values in the dataset. If there is any null value in any column, then fill it with the mean of that column


```python
 
 cars.isnull().sum()
 
```

```python
cars['Cylinders'].fillna(cars['Cylinders'].mean(),inplace = True)
cars['EngineSize'].fillna(cars['EngineSize'].mean(),inplace = True)
cars['Horsepower'].fillna(cars['Horsepower'].mean(),inplace = True)
cars['MPG_City'].fillna(cars['MPG_City'].mean(),inplace = True)
cars['Weight'].fillna(cars['Weight'].mean(),inplace = True)
cars['Wheelbase'].fillna(cars['Wheelbase'].mean(),inplace = True)
cars['Length'].fillna(cars['Length'].mean(),inplace = True)
cars['MPG_Highway'].fillna(cars['MPG_Highway'].mean(),inplace = True)
```

# Check what are the different types of Make are there in our dataset. And, what is the count (occurrence) of each Make in the data 

```python


cars['Make'].value_counts()
```

# ( Filtering ) - Show all the records where Origin is Asia or Europe.

```python


cars[cars['Origin'].isin(cars['Origin'] == 'Asia') | (cars['Origin'] == 'Europe')]
```

# ( Removing unwanted records ) - Remove all the records (rows) where Weight is above 4000.

```python

cars[~(cars['Weight'] > 4000)]
```

# ( Applying function on a column ) - Increase all the values of 'MPG_City' column by 3.

```python


cars['MPG_City'].apply(lambda x:x+3)
```
